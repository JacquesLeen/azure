differences between ```loc``` and ```iloc```:
* loc returns rows with index label between the provided list (0 to 5) with left extreme included
* iloc returns the rows with the position 0 to 5 

````python
# if I reorder the elements in the df iloc returns the first k according to the new order
df_students.sample(frac=1).iloc[0:5]
````

````python
df_students.iloc[0:4,[1,2]]
````

````python
#loc can be used to filter based on condition
df_students.loc[lambda df: df['Grade']>50]
````

````python
# the same can be achieved without loc
filter = df_students['Grade'] > 50
#this syntax generates a serie of booleans
print(filter)
print(type(filter))
````

````python
# and this can now be applied to filter the initial data
df_students[filter]
````

````python
# or again this can be done with query

df_students.query("Grade > 50")
````

## Read Data from File

````python
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
df_students = pd.read_csv('grades.csv',delimiter=',',header='infer')
df_students.head()
````

````python
#null values
# is null gets applied to all columns and returns a df of booleans
df_students.isnull()
````

````python
#far more interesting is this kind of report
df_students.isnull().sum()
````

````python
# or even better the filtering especially toward a column
filter = df_students['Grade'].isnull()
df_students[filter]
````

````python
#if I want to have any element that has a null somewhere then there is a more complex syntax

n_rows= 1000
n_columns = 5

df= pd.DataFrame( data=np.random.randn(n_rows, n_columns), columns=[f"col_{i}" for i in range(n_columns)])

null_ratio = 0.1
num_nulls = int(null_ratio*n_rows*n_columns)
for i in range(num_nulls):
    row= np.random.randint(0,n_rows)
    col = np.random.randint(0,n_columns)
    df.iat[row,col]=np.nan

df
````

````python
#this df has a certain number of null, but given that there are many rows an columns how do i identify them?

filter = df.isnull().any(axis=1)
df[filter]
````

````python
# one common approach would be to fill those values with the mean
print(df.median())
# df mean / median already evaluates per column, so
df = df.fillna(df.median())

df.isnull().sum()

````

````python
# we can do the same with our actual df
# we first select only the numerical columns
numerical = df_students.select_dtypes(include=[np.number])

# calculate the mean for those columns and fillna
numerical_filled = numerical.fillna(numerical.mean())

# change the columns
df_students[numerical.columns] =numerical_filled

df_students
````

````python
# select only students who studied more than average hours

avg_hours_study = df_students['StudyHours'].mean()

df_students.loc[lambda df: df['StudyHours']> avg_hours_study]
````

````python
# and what was their mean grade
df_students.loc[lambda df: df['StudyHours']> avg_hours_study]['Grade'].mean()
````

````python
# add column for pass / fail (grade >= 60)

passes = pd.Series(df_students['Grade']>=60)
df_students = pd.concat([df_students, passes.rename("Pass")], axis=1)
df_students
````

````python
# group data according to passing

df_students.groupby("Pass")['Name'].count()
````

````python
# or even better get informative indication of whether hours spent studying had an influence on passing the test

pd.DataFrame(df_students.groupby("Pass")["StudyHours"].mean().reset_index())
````
